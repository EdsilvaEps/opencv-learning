```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
# opencv converts the image into a numpy array automatically
img = cv2.imread('Computer-Vision-with-Python/DATA/00-puppy.jpg')
type(img)
```

```python
img.shape
```

```python
# opencv and matplotlib expect different order for the RGB pattern
# this causes the img to appear to have the wrong colors
plt.imshow(img)
```

```python
# MATPLOTLIB --> RBG RED GREEN BLUE
# OPENCV --> BLUE GREEN RED
# converting the opencv image color to a matplotlib rgb type
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
#  open the image using grayscale
img_gray = cv2.imread('Computer-Vision-with-Python/DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
# it doesnt have anymore the RGB dimension
img_gray.shape
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
# resizing the image using opencv
new_img = cv2.resize(fix_img,(1000,400))
plt.imshow(new_img)
```

```python
# resizing the image by ratio
w_ratio = 0.5
h_ratio = 0.5
new_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
# image is resized but not distorced 
plt.imshow(new_img)
```

```python
# flipping image accross the vertical axis
new_img = cv2.flip(fix_img, 0)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, -1)
plt.imshow(new_img)
```

```python
# saving our image 
cv2.imwrite('tottally_new.jpg', fix_img)
```

```python
# enlarging the matplotlib canvas
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.imshow(fix_img)
```
