```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Computer-Vision-with-Python/DATA/sammy_face.jpg')
```

```python
plt.imshow(img)
```

```python
# trying to use the algoritm straigh away with average values
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
plt.imshow(edges)
```

```python
# tweaking the threshold values
edges = cv2.Canny(image=img, threshold1=0, threshold2=255)
plt.imshow(edges)
```

```python
# calculating the median pixelvalues 
med_val = np.median(img)
med_val
```

```python
# LOWER THRESHOLD TO EITHER 0 OR 70% OF THE MEDIAN VALUE, WHICHEVER IS GREATER
lower = int(max(0,0.7*med_val))
# UPPER THRESHOLD TO EITHER 130% OF THE MEDIAN OR THE MAX 255, WHICHEVER IS SMALLER
upper = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(image=img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
# blurring the image for better edged detection
blurred_img = cv2.blur(img,ksize=(5,5))
edges = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
# trying with ever stronger blue
blurred_img = cv2.blur(img,ksize=(7,7))
edges = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
# trying with higher threshold
blurred_img = cv2.blur(img,ksize=(5,5))
edges = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper+50)
plt.imshow(edges)
```
