## BLENDING IMAGES OF THE SAME SIZE

```python
# resizing the two images above so they're the same size
img1 = cv2.resize(img1,(1200,1200))
img2 = cv2.resize(img2,(1200,1200))
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
# blending the images
# alpha and beta are variables to identify which of the images do show more
# only works on images the same size
blended = cv2.addWeighted(src1=img1, alpha=0.8, src2=img2, beta=0.3, gamma=0)
plt.imshow(blended)
```

# MASKS

## BLEND TOGETHER IMAGES OF DIFFERENT SIZES

```python
# importing the images and changin their color mapping
img1 = cv2.imread('Computer-Vision-with-Python/DATA/dog_backpack.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('Computer-Vision-with-Python/DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2 = cv2.resize(img2,(600,600))
```

```python
# choosing ROI (region of interest)
plt.imshow(img1)
```

```python
img1.shape
```

```python
# setting ROI to the bottom-right of the image
x_offset = 934 - 600
y_offset = 1401 - 600
```

```python
rows,cols,channels = img2.shape
```

```python
# selecting the ROI
roi = img1[y_offset:1401, x_offset:943]
plt.imshow(roi)
```

```python
# creating a mask
img2gray = cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
plt.imshow(img2gray, cmap='gray')
```

```python
# we want to make the areas of interest 'white'
mask_inv = cv2.bitwise_not(img2gray) # inverts the colors of the image
plt.imshow(mask_inv, cmap='gray')
```

```python
# the new image has no color channel anymore, they need to be added again
mask_inv.shape
```

```python
import numpy as np

white_background = np.full(img2.shape, 255,dtype=np.uint8)
```

```python
white_background.shape
```

```python
# this command takes the white_background img and puts the mask_inv over it
bk = cv2.bitwise_or(white_background,white_background,mask=mask_inv)
bk.shape
```

```python
plt.imshow(bk)
```

```python
# put the mask over the original img
fg = cv2.bitwise_or(img2, img2, mask=mask_inv)
plt.imshow(fg)
```

```python
# paste fg on the roi
final_roi = cv2.bitwise_or(roi,fg)
plt.imshow(final_roi)
```

```python
# overlaying final_roi on the original image again
large_img, small_img = img1, final_roi
large_img[y_offset:y_offset+small_img.shape[0], x_offset:x_offset+small_img.shape[1]] = small_img
plt.imshow(large_img)
```
