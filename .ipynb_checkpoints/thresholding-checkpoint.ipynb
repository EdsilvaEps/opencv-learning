```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read images as greyscale
img = cv2.imread('Computer-Vision-with-Python/DATA/rainbow.jpg',0)
plt.imshow(img, cmap='gray')
```

```python
# thesholding an image (using binary thresholding)
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret
```

```python
# thresheld image (all values are either 0 or 255)
plt.imshow(thresh1, cmap='gray')
```

```python
# thesholding an image (using inverse binary thresholding)
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh1, cmap='gray')
```

```python
# thesholding an image (using threshold truncated)
# if a pixel has value higher than max allowed, TRUNC brings it down to this values
# else it leaves it be
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
plt.imshow(thresh1, cmap='gray')
```

```python
# thesholding an image (using threshold truncated inverse)
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
plt.imshow(thresh1, cmap='gray')
```

```python
img = cv2.imread('Computer-Vision-with-Python/DATA/crossword.jpg',0)
plt.imshow(img,cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
# quality lost with binary threshold
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
show_pic(th1)
```

```python
# use adpative threshold to get the most of the picture
th2= cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,8)
show_pic(th2)
```

```python
# its still not so nice, lets blend the two previous images
# it looks better now!
blended = cv2.addWeighted(src1=th1, alpha=0.6, src2=th2, beta=0.4, gamma=0)
show_pic(blended)
```
