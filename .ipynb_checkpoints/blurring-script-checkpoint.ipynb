```python
# importing test image
path = 'Computer-Vision-with-Python/DATA/'
img = cv2.imread(path+'car_plate.jpg')
plt.imshow(img)
```

```python
# function that displays image in larger scale and with correct coloring
def display(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
display(img)
```

```python
# loading a classifier (Haar Cascade) pretrained for identifying russian plate numbers
plates_cascade = cv2.CascadeClassifier(path+'haarcascades/haarcascade_russian_plate_number.xml')
```

```python
# function that detects the russian license plates and draws a rectangle around them
# returns the processed image
def detect_plate(img):
    license_img = img.copy()
    license_rects = plates_cascade.detectMultiScale(img, scaleFactor=1.2, minNeighbors=3)
    
    for (x,y,w,h) in license_rects:
        cv2.rectangle(license_img, (x,y), (x+w,y+h), (0, 0, 255), 10)
        
    return license_img
```

```python
result = detect_plate(img)
display(result)
```

```python
# function to detect and blur the license plates of cars
def detect_and_blur_plate(image):
    license_img = image.copy() # making a copy of the image to edit
    # the following function returns a list of rectangles encompassing the detected instances of objects
    license_rects = plates_cascade.detectMultiScale(img, scaleFactor=1.2, minNeighbors=3)
    
    for (x,y,w,h) in license_rects:
        
        # grabbing the ROI (Region of Interest) from the images
        roi = license_img[y:y+h, x:x+w]
        roi = cv2.medianBlur(roi, 25) # blurring this particular piece of the image
        # overlaying the blurred piece back into the larger image
        license_img[y:y+roi.shape[0], x:x+roi.shape[1]] = roi
        
    return license_img
    
```

```python
result = detect_and_blur_plate(img)
display(result)
```
