```python
import numpy as np 
import matplotlib.pyplot as plt
# to display plots inside the notebook
%matplotlib inline
```

```python
from PIL import Image
```

```python
# opening a picture with PIL
pic = Image.open('Computer-Vision-with-Python/DATA/00-puppy.jpg')
type(pic)
```

```python
# turning this picture into an array
pic_arr = np.asarray(pic)
type(pic_arr)
```

```python
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy()
# selecting only one color channel(red) in the image
# R G B 
# showing the red channel values(0-255) in grayscale
plt.imshow(pic_red[:,:,0], cmap='gray')
```

```python
# showing the green channel values(0-255) in grayscale
plt.imshow(pic_red[:,:,1], cmap='gray')
```

```python
# showing the blue channel values(0-255) in grayscale
plt.imshow(pic_red[:,:,2], cmap='gray')
```

```python
# setting the green channel to zero
pic_red[:,:,1] = 0
plt.imshow(pic_red)
# then only the colors left and blue and red, that makes purple
```

```python
# setting the blue channel to zero
pic_red[:,:,2] = 0
plt.imshow(pic_red)
# only red remains
```
