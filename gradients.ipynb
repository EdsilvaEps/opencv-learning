```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# an image gradient is a directional change in the intensity or color in an image
# some functions :
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = cv2.imread('Computer-Vision-with-Python/DATA/sudoku.jpg', 0)
display_img(img)
```

```python
# identifying the vertical lines
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
display_img(sobelx)
```

```python
# identifying the horizontal lines
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
display_img(sobely)
```

```python
# laplacian gradient calculation
laplacian = cv2.Laplacian(img, cv2.CV_64F)
display_img(laplacian)
```

```python
# blending both the sobel processed images
blended = cv2.addWeighted(src1=sobelx, alpha=0.5,src2=sobely,beta=0.5,gamma=0.5)
display_img(blended)
```

```python
# thresholding
ret,th1 = cv2.threshold(img,100,255,cv2.THRESH_BINARY)
display_img(th1)
```

```python
kernel = np.ones((4,4), np.uint8)
gradient = cv2.morphologyEx(blended, cv2.MORPH_GRADIENT, kernel)
display_img(gradient)
```
